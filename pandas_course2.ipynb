```python
# Python program to concatenate (union with columns as key)
# dataframes with same columns, different rows 

# Creating first dataframe 
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3'], 
                    'C': ['C0', 'C1', 'C2', 'C3'], 
                    'D': ['D0', 'D1', 'D2', 'D3']}, 
                    index = [0, 1, 2, 3]) 
df1
```

```python
# Creating second dataframe  3 rows
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6'], 
                    'B': ['B4', 'B5', 'B6'], 
                    'C': ['C4', 'C5', 'C6'], 
                    'D': ['D4', 'D5', 'D6']}, 
                    index = [4, 5, 6]) 
df2
```

```python
# Creating third dataframe  3 columns
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'], 
                    'B': ['B8', 'B9', 'B10', 'B11'], 
                    'C': ['C8', 'C9', 'C10', 'C11'] 
                    }, 
                    index = [8, 9, 10, 11]) 
df3
```

```python
# Concatenating the dataframes with different number of columns or different number of rows
df=pd.concat([df1, df2, df3], sort=False) 
df
```

```python
# Python program to merge  (union with intersection key)  
# dataframes with same rows, different columns and same key column
# Dataframe created 
left = pd.DataFrame({'Key': ['K0', 'K1', 'K2', 'K3'], 
                    'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3']}) 
left
```

```python
right = pd.DataFrame({'Key': ['K0', 'K1', 'K2', 'K3'],   #same column
                    'C': ['C0', 'C1', 'C2', 'C3'], 
                    'D': ['D0', 'D1', 'D2', 'D3']}) 
right
```

```python
# Merging the dataframes                   
df=pd.merge(left, right, how ='inner', on ='Key') 
df
```

```python
# Python program to join (union)
# dataframes with different columns, different rows

left = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
					'B': ['B0', 'B1', 'B2', 'B3']}, 
					index = ['K0', 'K1', 'K2', 'K3']) 

right = pd.DataFrame({'C': ['C0', 'C1', 'C2', 'C3'], 
					'D': ['D0', 'D1', 'D2', 'D3']}, 
					index = ['K0', 'K1', 'K2', 'K3']) 
						
# Joining the dataframes 
left.join(right) 
```
